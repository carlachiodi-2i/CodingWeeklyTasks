```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import requests, io
import matplotlib.pyplot as plt 


searchHelloWorld = "Hello World"
num_requested = 1
images = []

searchWithoutSpaces = searchHelloWorld.replace(" ", "+").lower()
url = "https://www.google.com/search?q="+searchWithoutSpaces+"&source=lnms&tbm=isch"
header={'User-Agent':"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.134 Safari/537.36"
}
soup = BeautifulSoup(urlopen(Request(url,headers=header)),'html.parser')
for image in soup.findAll('img'):
    image_src = image.get('src')
    if image_src.startswith('https'):
        images.append(image_src)


response = requests.get(images[2]).content
img = plt.imread(io.BytesIO(response), format='JPG')
plt.yticks([])
plt.xticks([])
plt.imshow(img)
```
